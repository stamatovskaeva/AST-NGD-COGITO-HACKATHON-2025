```python
import cv2
import numpy as np

cap = cv2.VideoCapture('videos/Nesten alle varer sakte tempo 480P.mp4')
ret, prev_frame = cap.read()
prev_frame_gray = cv2.cvtColor(prev_frame, cv2.COLOR_BGR2GRAY)

event_triggered = False
event_threshold = 50000  # Adjust this threshold based on your experiment

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break
        
    frame_gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    
    # Compute the absolute difference between current frame and previous frame
    diff = cv2.absdiff(frame_gray, prev_frame_gray)
    
    # Threshold the difference to get a binary image
    _, diff_thresh = cv2.threshold(diff, 30, 255, cv2.THRESH_BINARY)
    
    # Calculate the sum of pixel values in the thresholded image
    diff_sum = np.sum(diff_thresh)
    
    # Check if the sum exceeds the threshold to trigger an event
    if diff_sum > event_threshold:
        print("Event triggered!")
        # You could run your object detection or tracking here
        event_triggered = True
    else:
        event_triggered = False

    # Optionally display the difference image for debugging
    cv2.imshow("Diff", diff_thresh)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
    
    # Update the previous frame
    prev_frame_gray = frame_gray

cap.release()
cv2.destroyAllWindows()
```
