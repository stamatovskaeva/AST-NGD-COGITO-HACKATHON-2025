```python
import cv2

cap = cv2.VideoCapture('videos/Nesten alle varer sakte tempo 480P.mp4')
fgbg = cv2.createBackgroundSubtractorMOG2(history=500, varThreshold=50, detectShadows=True)

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break
    
    fgmask = fgbg.apply(frame)
    
    # Optional: apply a threshold to remove shadows or minor noise
    _, fgmask = cv2.threshold(fgmask, 250, 255, cv2.THRESH_BINARY)
    
    # Count the number of white pixels (i.e., moving regions)
    movement = cv2.countNonZero(fgmask)
    
    if movement > 1000:  # Adjust the threshold based on your scene
        print("Significant movement detected: Event triggered!")
        # Run detection/tracking as needed here
    
    cv2.imshow('Foreground Mask', fgmask)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
